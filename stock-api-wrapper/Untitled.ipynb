```python
data = pd.read_csv("stocks.csv").iloc[:, :2]
data.columns = [c.lower() for c in data.columns]
data
```

```python
import string

def clean(name):
    """
    remove punctuation and convert to lowercase
    """
    out = "".join([ch for ch in name.lower() if ch not in string.punctuation])
    return out

data["name2"] = [clean(name) for name in data["name"]]
data["name3"] = [name.replace(" ", "") for name in data["name2"]]

data
```
